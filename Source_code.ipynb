#### Importing Modules

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt 
```

#### Importing Dataset

```python
data_set= pd.read_csv('datasets/students_scores.csv')  

x= data_set.iloc[:, :-1].values  
y= data_set.iloc[:, 1].values   
```

#### Defining Model

```python
from sklearn.model_selection import train_test_split  
x_train, x_test, y_train, y_test= train_test_split(x, y, test_size= 1/3, random_state=0) 


from sklearn.linear_model import LinearRegression  
model= LinearRegression()  
model.fit(x_train, y_train)  

y_pred= model.predict(x_test)  
x_pred= model.predict(x_train) 
```

#### Plotting the Dataset

```python
plt.scatter(x_train, y_train, color="green")   
plt.plot(x_train, x_pred, color="red")    
plt.title("HOURS VS STUDY (Training Dataset)")  
plt.xlabel("hours")  
plt.ylabel("study")  
plt.show()   


print(model.coef_)
print(model.intercept_)
```

#### Predicting the score for 9.5 Hrs of study

```python
def calc(slope, intercept, hours):
    return slope*hours+intercept
score = calc(model.coef_, model.intercept_, 9.5)
hrs = np.array([[9.5]])
score = model.predict(hrs)
print("The score of the student:",score.squeeze())
```
